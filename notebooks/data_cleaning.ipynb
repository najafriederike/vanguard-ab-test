```python
import pandas as pd

def clean_and_merge_data(demo_path, exp_clients_path, web_data_path):
    """
    Cleans data by dropping rows with missing values important columns, and convert to datetime format.

    Parameters:
    - demo_path (str): Path to the demographic data CSV file.
    - exp_clients_path (str): Path to the experiment clients CSV file.
    - web_data_path (str): Path to the web data CSV file.

    Returns:
    - pd.DataFrame: Cleaned and merged DataFrame.
    """
    # Load the datasets
    df_demo = pd.read_csv(demo_path)
    df_exp_clients = pd.read_csv(exp_clients_path)
    df_web_data = pd.read_csv(web_data_path)
    
    # Step 1: Clean demographic data
    # Drop rows with missing values in key columns
    df_demo_cleaned = df_demo.dropna(subset=['clnt_tenure_yr', 'clnt_tenure_mnth', 'clnt_age', 'gendr'])
    
    # Step 2: Clean experiment clients data
    # Drop rows with missing Variation values
    df_exp_clients_cleaned = df_exp_clients.dropna(subset=['Variation'])
    
    # Step 3: Clean web data
    # Convert date_time to datetime format
    df_web_data['date_time'] = pd.to_datetime(df_web_data['date_time'])
    
    # Step 4: Merge demographic and experiment data
    df_merged = pd.merge(df_demo_cleaned, df_exp_clients_cleaned, on='client_id', how='inner')
    
    # Step 5: Merge the result with web data
    df_final = pd.merge(df_merged, df_web_data, on='client_id', how='inner')
    
    # Return cleaned and merged DataFrame
    return df_final
```

```python
# File paths to the datasets
demo_path = '../data/clean/df_final_demo.csv'
exp_clients_path = '../data/clean/df_final_exp_clients.csv'
web_data_path = '../data/clean/df_final_web_data_merged.csv'

# Call the cleaning function
df_cleaned_final = clean_and_merge_data(demo_path, exp_clients_path, web_data_path)

# Inspect the cleaned and merged data
print(df_cleaned_final.info())
print(df_cleaned_final.head())
```
